```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
url = "https://raw.githubusercontent.com/dsrscientist/Dataset2/main/temperature.csv"
data = pd.read_csv(url)

# Count the occurrences of each station
station_counts = data['station'].value_counts()

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(station_counts.index, station_counts.values, color='blue')
plt.xlabel('Station')
plt.ylabel('Number of Data Points')
plt.title('Distribution of Data across Weather Stations')
plt.xticks(rotation=45)
plt.grid(axis='y')

plt.show()
```

```python


# Convert 'Date' column to datetime
data['Date'] = pd.to_datetime(data['Date'])

# Count the occurrences of each date
date_counts = data['Date'].value_counts().sort_index()

# Create a time series plot
plt.figure(figsize=(12, 6))
plt.plot(date_counts.index, date_counts.values, marker='o', color='blue', linestyle='-', linewidth=2, markersize=6)
plt.xlabel('Date')
plt.ylabel('Number of Data Points')
plt.title('Distribution of Data across Dates')
plt.xticks(rotation=45)
plt.grid()

plt.show()
```
